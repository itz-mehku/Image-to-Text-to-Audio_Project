## **--> Image to Text to Audio Generation using Tesseract OCR and Google Text-to-Speech**

```python
import pytesseract
from PIL import Image
import gtts
```

```python
img = Image.open('/content/Picture1 (1).png') # path of the image
print(img)
```

```python
result = pytesseract.image_to_string(img) # convert image to text
print(result)
tts = gtts.gTTS(result)
tts.save("hello.mp3")
```
